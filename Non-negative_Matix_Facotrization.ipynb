```python
def read_nodeadjlist(filename):
  ego = int(filename.rsplit('/',1)[1].split('.')[0])
  G = nx.Graph()
  for line in open(filename):
    e1, es = line.split(':')
    G.add_edge(ego,int(e1))
    es = es.split()
    for e in es:
      if e == e1: continue
      G.add_edge(int(e1),int(e))
  return G
```

```python
def naive_spec_cluster(G,k=8):
    cl = sklearn.cluster.spectral_clustering(nx.adjacency_matrix(G),n_clusters=k)
    pred_circles ={}
    for circle,user in zip(cl,G.nodes()):  # ordering is the same as G.nodes()
        if circle in pred_circles:
          pred_circles[circle].append(user)
        else:
          pred_circles[circle]=[user]
    return pred_circles
```

```python
def compute_training_score(cluster_function):
    trainingfiles = os.listdir('./Training/')
    pred_score = {}
    
    for item in trainingfiles:
        ego = int((item.split('.')[0]))
        true_circles = readcirclefile('./Training/'+item)
        G = read_nodeadjlist('./egonets/'+str(ego)+'.egonet')
        pred_circles = cluster_function(G)
        for key,val in pred_circles.items():
                if ego in val:
                    pred_circles[key].remove(ego)
        pred_score[ego] = cost_function(pred_circles,true_circles)
    return pred_score
```

```python
def NMF_known_cluster(ego,beta=0.1):
    G = read_nodeadjlist('./egonets/'+str(ego)+'.egonet')
    true_circles = readcirclefile('./Training/'+str(ego)+'.circles')
    N_comp = len(true_circles)
    
    L = -nx.laplacian_matrix(G)
    expL = sp.linalg.expm(beta * L)
    
    N_nodes = len(G.nodes())
    B = np.zeros([N_nodes,N_nodes])
    
    for i in range(N_nodes):
        for j in range(N_nodes):
            B[i,j] = expL[i,j]/np.sqrt(expL[i,i]*expL[j,j])
    
            
    nmf = sk.decomposition.NMF(n_components=N_comp)
    comps= nmf.fit_transform(B)
    
    minthresh,maxthresh,stepthresh = 0.05,0.95,0.01
    thresh_vals = arange(minthresh,maxthresh,stepthresh)
    score_v_thresh = np.zeros([len(thresh_vals),2])
                               
    for k,thresh in enumerate(thresh_vals):
        pred_circles = dict(zip(range(N_comp),[[] for _ in range(N_comp)]))
        node_list = G.nodes()
        for i in range(N_comp):
            for j,val in enumerate(comps[:,i]):
                if val>thresh and node_list[j] != ego:
                    pred_circles[i].append(node_list[j])
        score_val = cost_function(pred_circles,true_circles)
        score_v_thresh[k,0]=thresh
        score_v_thresh[k,1]=score_val
    
    return min(score_v_thresh,key=lambda x: x[1])
        
```

```python
def convert_circles_to_partition(circles):
    return dict(list(itertools.chain(*[[(user,label) for user in circle] for label, circle in circles.items()])))
```

```python
def convert_nested_circles_to_partition(circles,G):
    friends = set(G.nodes())
    out = dict()
    for label,circle in circles.items():
        newcircle = set(circle)
        for plabel,pcircle in out.items():
            newcircle = newcircle.difference(set(pcircle))
        if len(newcircle)>0:
            out[label] = list(newcircle)
    no_circle = list(friends.difference(*map(set,out.values())))
    if len(no_circle)>0:
        out['no_circle'] = no_circle
    return out
```

```python
def modularity_overlap(circles,G):
    G_nodes = G.nodes()
    N_nodes = len(G_nodes)
    m = len(G.edges())
    A=nx.adjacency_matrix(G)
    
    S = np.zeros([N_nodes,N_nodes])
    
    gset = []
    for i in range(N_nodes):
        gs = [set(circle) for circle in circles.values() if G_nodes[i] in circle]
        if len(gs)>0:
            gset.append( set.union(*gs) )
        else:
            gset.append(set())
            
    for i in range(N_nodes):
        for j in range(N_nodes):
            if i!=j and len(gset[i])>0 and len(gset[j])>0:
                S[i,j] = (len(gset[i].intersection(gset[j]))/(1.*len(gset[i].union(gset[j]))))
       
    asum = 0
    for i in range(N_nodes):
        for j in range(N_nodes):
            if i!=j:           
                asum += (A[i,j] - G.degree(G_nodes[i])*G.degree(G_nodes[j])/(2.*m))*S[i,j]
    
    return asum/(2.*m)
```

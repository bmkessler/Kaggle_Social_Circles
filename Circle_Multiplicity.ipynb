```python
trainingfiles = os.listdir('./Training/')

# compute the number of clusters based on the weak fit vs. friend number
m = 0.01670827551
b = 5.82354246449

all_in_one = {}
con_comp = {}
clust_score = {}
nfclust_score = {}

for item in trainingfiles:
    ego = int((item.split('.')[0]))
    circles = readcirclefile('./Training/'+item)
    G = read_nodeadjlist('./egonets/'+str(ego)+'.egonet')
    num_friends = len(G.nodes())
    n_clust = int(m*num_friends+b)
    all_in_one[ego] = cost_function({0:G.nodes()},circles)
    con_comp[ego] = cost_function(dict(enumerate(nx.connected_components(G))),circles)
    cl = sklearn.cluster.spectral_clustering(nx.adjacency_matrix(G))
    nfcl = sklearn.cluster.spectral_clustering(nx.adjacency_matrix(G),n_clusters=n_clust)
    pred_circles ={}
    for circle,user in zip(cl,G.nodes()):  # if no node_list is given to adjacency_matrix the ordering is the same as G.nodes()
        if circle in pred_circles:
            pred_circles[circle].append(user)
        else:
            pred_circles[circle]=[user]
    clust_score[ego] = cost_function(pred_circles,circles)
    pred_circles ={}
    for circle,user in zip(nfcl,G.nodes()):  # if no node_list is given to adjacency_matrix the ordering is the same as G.nodes()
        if circle in pred_circles:
            pred_circles[circle].append(user)
        else:
            pred_circles[circle]=[user]
    nfclust_score[ego] = cost_function(pred_circles,circles)


print 'All-in-one error:', sum(all_in_one.values())
print 'Connected components error:', sum(con_comp.values())
print 'Naive (k=8) clustering error:', sum(clust_score.values())
print 'Weak N-friends (k = m*N + b) cluster error:', sum(nfclust_score.values())
```

```python
def naive_spec_cluster(G):
    cl = sklearn.cluster.spectral_clustering(nx.adjacency_matrix(G))
    pred_circles ={}
    for circle,user in zip(cl,G.nodes()):  # ordering is the same as G.nodes()
        if circle in pred_circles:
          pred_circles[circle].append(user)
        else:
          pred_circles[circle]=[user]
    return pred_circles
```

```python
def generate_class_matrix(G,true_circles):
    return [[int(g in circle) for g in G.nodes()] for circle in true_circles.values()]
```

```python
def detailed_match_vector(detailed_ego_profile, detailed_user_profile):
    """ compares direct match between ego profile features
    should be used with full:feature:value"""
    return [int(val in detailed_user_profile) for val in detailed_ego_profile]
```

```python
def read_profiles(featurefile):
    """
    reads a featurefile consisting of userid feature;value feature;value
    returns a dictionary keyed by userId, elements are lists of 'feature1;feature2;value'
    """
    with open(featurefile) as f:
        out = {} 
        for line in f:
            tokens = line.split()
            ego = int(tokens[0])
            out[ego] = sorted(tokens[1:])
        for i in range(len(out)):
            assert ('id;'+str(i)) in out[i]  # check that each line was read and placed in the correct place in the dictionary
    return out
```

```python
profiles_dict = read_profiles('features.txt')
```

```python
def generate_feature_matrix(profiles_dict,ego,G):
    return [detailed_match_vector(profiles_dict[ego], profiles_dict[g]) for g in G.nodes()]
```
